```python
# dependencies
import pandas as pd
import requests

# import all weather csvs
noaa2013 = "./Data/NOAA_2013.csv"
noaa2014 = "./Data/NOAA_2014.csv"
noaa2015 = "./Data/NOAA_2015.csv"
noaa2016 = "./Data/NOAA_2016.csv"
noaa2017 = "./Data/NOAA_2017.csv"
noaa2018 = "./Data/NOAA_2018.csv"
noaa2019 = "./Data/NOAA_2019.csv"
noaa2020 = "./Data/NOAA_2020.csv"

# read and store the files in dataframes
weather2013 = pd.read_csv(noaa2013)
weather2014 = pd.read_csv(noaa2014)
weather2015 = pd.read_csv(noaa2015)
weather2016 = pd.read_csv(noaa2016)
weather2017 = pd.read_csv(noaa2017)
weather2018 = pd.read_csv(noaa2018)
weather2019 = pd.read_csv(noaa2019)
weather2020 = pd.read_csv(noaa2020)

# combine the years of data into one dataframe
weatherData = pd.concat([weather2013,weather2014,weather2015,weather2016,weather2017,
                        weather2018,weather2019,weather2020])

# reset the index to avoid repeats
weatherData.reset_index(inplace=True,drop=True)
```

```python
# remove extraneous columns
weatherData = weatherData[["STATION","NAME","LATITUDE","LONGITUDE","DATE","DT32","DX90",
                          "EMXP","EMXT","HX01","PRCP","TAVG","TMAX","TMIN","WSF2","WSFG"]]

# rename columns
weatherData = weatherData.rename(columns={"STATION":"Station ID","NAME":"Station Name","LATITUDE":"Lat",
                "LONGITUDE":"Long","DATE":"Year"})
```

```python
weatherData
```

```python
# drop any rows that have all NaNs for weather data columns
weatherData = weatherData.dropna(subset=["DT32","DX90","EMXP","EMXT","HX01","PRCP",
                                         "TAVG","TMAX","TMIN","WSF2","WSFG"],how="all")
# reset the index
weatherData.reset_index(inplace=True,drop=True)

# add blank column for county
weatherData["County"]=""
weatherData = weatherData[["Station ID","Station Name","County","Lat","Long","Year","DT32","DX90",
                          "EMXP","EMXT","HX01","PRCP","TAVG","TMAX","TMIN","WSF2","WSFG"]]
weatherData
```

```python
# use the census API to find each entry's county using coordinates
# build query url
base_url = "https://geo.fcc.gov/api/census/block/find?"

# loop through dataframe to find county for each lat long
for index, row in weatherData.iterrows():
    
    latitude = row["Lat"]
    longitude = row["Long"]
    
    query_url = (f"{base_url}latitude={latitude}&longitude={longitude}&showall=true&format=json")
    
    response = requests.get(query_url)
    response = response.json()
#     results = response["results"]
    
    print(f"Finding County for index {index}")
    weatherData.loc[index,"County"]=response["County"]["name"]
    
weatherData
```

```python
dt32 = weatherData[["Station ID","Station Name","County","Lat","Long","Year","DT32"]]
dx90 = weatherData[["Station ID","Station Name","County","Lat","Long","Year","DX90"]]
emxp = weatherData[["Station ID","Station Name","County","Lat","Long","Year","EMXP"]]
emxt = weatherData[["Station ID","Station Name","County","Lat","Long","Year","EMXT"]]
hx01 = weatherData[["Station ID","Station Name","County","Lat","Long","Year","HX01"]]
prcp = weatherData[["Station ID","Station Name","County","Lat","Long","Year","PRCP"]]
tavg = weatherData[["Station ID","Station Name","County","Lat","Long","Year","TAVG"]]
tmax = weatherData[["Station ID","Station Name","County","Lat","Long","Year","TMAX"]]
tmin = weatherData[["Station ID","Station Name","County","Lat","Long","Year","TMIN"]]
wsf2 = weatherData[["Station ID","Station Name","County","Lat","Long","Year","WSF2"]]
wsfg = weatherData[["Station ID","Station Name","County","Lat","Long","Year","WSFG"]]
```

```python
# drop NaNs for each dataframe
dt32 = dt32.dropna()
dx90 = dx90.dropna()
emxp = emxp.dropna()
emxt = emxt.dropna()
prcp = prcp.dropna()
tavg = tavg.dropna()
tmax = tmax.dropna()
tmin = tmin.dropna()


# hx01 only has 7 data points - removing this variable from the analysis
hx01 = hx01.dropna()

# wsfg only has 2 data points - removing this variable from the analysis
wsfg = wsfg.dropna()

# wsf2 only has 540 data points - removing this variable from the analysis
wsf2 = wsf2.dropna()

# new master dataframe without these variables
weatherData = weatherData[["Station ID","Station Name","County","Lat","Long","Year",
                           "DT32","DX90","EMXP","EMXT","PRCP","TAVG","TMAX","TMIN"]]
```

```python
# group each dataframe by year and county and aggregate the weather column appropriately
# aggregation type is dependent on the variable

# DT32 - average days per county
dt32_grouped = dt32.groupby(["Year","County"])
dt32_CountyMean = dt32_grouped["DT32"].mean()
pd.DataFrame(dt32_CountyMean)

# DX90 - average days per county
dx90_grouped = dx90.groupby(["Year","County"])
dx90_CountyMean = dx90_grouped["DX90"].mean()
pd.DataFrame(dx90_CountyMean)

# emxp - maximum inches per county
emxp_grouped = emxp.groupby(["Year","County"])
emxp_CountyMax = emxp_grouped["EMXP"].max()
pd.DataFrame(emxp_CountyMax)

# emxt - maximum temperature per county
emxt_grouped = emxt.groupby(["Year","County"])
emxt_CountyMax = emxt_grouped["EMXT"].max()
pd.DataFrame(emxt_CountyMax)

# prcp - average inches per county
prcp_grouped = prcp.groupby(["Year","County"])
prcp_CountyMean = prcp_grouped["PRCP"].mean()
pd.DataFrame(prcp_CountyMean)

# tavg - average temperature per county
tavg_grouped = tavg.groupby(["Year","County"])
tavg_CountyMean = tavg_grouped["TAVG"].mean()
pd.DataFrame(tavg_CountyMean)

# tmax - average temperature per county
tmax_grouped = tmax.groupby(["Year","County"])
tmax_CountyMean = tmax_grouped["TMAX"].mean()
pd.DataFrame(tmax_CountyMean)

# tmin - average temperature per county
tmin_grouped = tmin.groupby(["Year","County"])
tmin_CountyMean = tmin_grouped["TMIN"].mean()
pd.DataFrame(tmin_CountyMean)
```

```python
# export to csvs
weatherData.to_csv("./Clean Data/WeatherData_Clean.csv")
dt32_CountyMean.to_csv("./Clean Data/DT32_Clean.csv")
dx90_CountyMean.to_csv("./Clean Data/DX90_Clean.csv")
emxp_CountyMax.to_csv("./Clean Data/EMXP_Clean.csv")
emxt_CountyMax.to_csv("./Clean Data/EMXT_Clean.csv")
prcp_CountyMean.to_csv("./Clean Data/PRCP_Clean.csv")
tavg_CountyMean.to_csv("./Clean Data/TAVG_Clean.csv")
tmax_CountyMean.to_csv("./Clean Data/TMAX_Clean.csv")
tmin_CountyMean.to_csv("./Clean Data/TMIN_Clean.csv")
```
